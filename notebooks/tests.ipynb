```python
import numpy as np
from pybats.dglm import pois_dglm
```

```python
deltrend = 1.0
a0 = np.array([[3]]) # prior state vector mean
R0= np.diag([0.01]) # prior state vector covariance


mod = pois_dglm(a0=a0,
                R0=R0,
                ntrend=1, # 1 ntrend terms gives an intercept
                deltrend=deltrend) # Discount factor on the two trend terms
```

```python
l = 1000
sim = np.zeros(l)
for i in range(l):
    sim[i] = mod.forecast_marginal(k=1)
    mod.update(sim[i])
```
